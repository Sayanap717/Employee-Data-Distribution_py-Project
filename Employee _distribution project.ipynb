#### ðŸ‘‡ Determine the distribution of employees across each team and calculate the percentage split relative to the total number of employees

```python
df_count=df["Team"].value_counts()
df_count
```

### ðŸ‘‰ distrbuton of employees in each team

```python
df.info()
```

```python
df_percentage= (df_count/len(df))*100
df_percentage
```

 #### ðŸ‘‡Barplot represents the number of employess in each team and the percentage split relative to the total number of employees

```python
plt.figure(figsize=(20,10))
sns.barplot(x=df_percentage.index,y=df_percentage.values, palette='light:#5A9',)
plt.xlabel("Team Name",fontsize=15)           #xlabel
plt.ylabel("Employee Count",fontsize=15)      #ylabel
plt.xticks(rotation=90)                       #vertical printing of x axis
plt.title("Distribution of Employees across each Team",fontsize=25)
for i,percentage in enumerate(df_percentage):
    plt.text(i, percentage + 0.1, f'{df_percentage[i]:.2f}%', ha='center', va='top')
plt.show()
```

#### ðŸ‘‡ Identify the predominant age group among employees. 

```python
df_preage= df["Age"].value_counts()
df_preage
```

```python
bins = [18, 24, 34, 44, 54, 65]
labels = ['18-24', '25-34', '35-44', '45-54', '55-65']
df["Age Group"] = pd.cut(df["Age"], bins=bins, labels=labels, right=False)  # Categorize the ages into age groups
df["Age Group"]
```

```python
predominant_agegroup = df["Age Group"].value_counts()    #Find the predominant age group
print(predominant_agegroup)
```

âœ… The predominant age group range is 25-34

```python
plt.figure(figsize=(10,5))
plt.plot(predominant_agegroup.index,predominant_agegroup.values,marker='o',linestyle='dotted',color='blue',markersize=10)
plt.title('Employee Distribution by Age Group',fontweight='bold',fontsize=12,color='orange')
plt.xlabel("Age Group",fontsize=10)           #xlabel
plt.ylabel("Count of Employees",fontsize=10)      #ylabel
plt.show()
```

#### Discover which team and position have the highest salary expenditure. 

```python
df
```

```python
df["Salary"].max()
```

```python
df["Salary"] = pd.to_numeric(df["Salary"], errors='coerce') # convert to numeric value
```

```python
df[df["Salary"]==df["Salary"].max()]
```

```python
salary_expenditure = df.groupby(['Team', 'Position'])['Salary'].sum().unstack()
salary_expenditure
```

```python
salary_expenditure.plot(kind='bar',figsize=(15,8), stacked=True)
plt.title('Salary Expenditure by Position and Team',fontweight ='bold',fontsize = 15)
plt.xlabel('Team')
plt.ylabel('Total Salary Expenditure')
plt.xticks(rotation=90)
plt.show()
```

```python
correlation_matrix = df[['Age', 'Salary']].corr()
correlation_matrix
```

## ðŸ‘† Correlation between Age and Salary

```python
plt.figure(figsize=(6,6))
sns.heatmap(correlation_matrix, annot=True, cmap='RdBu', center=0)
plt.title('Correlation Heatmap Between Age and Salary')
plt.show()
```
